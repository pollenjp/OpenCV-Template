```python
#===============================================================================
# Find contours(輪郭)
contours, hierarchy = cv2.findContours(image=im,
                                       mode=cv2.RETR_TREE,
                                       method=cv2.CHAIN_APPROX_SIMPLE,
                                      )

print("len(contours)     : {}".format(len(contours)))
for i, contour in enumerate(contours):
    print("contours[{:2}] | area : {:>7} | shape : {:}".format(i, cv2.contourArea(contour=contour), contour.shape))
    if i > 10:
        break
```

### 短形の抽出

#### bounding

- Bounding Rectangle
- Rotated Rectangle

```python
#===============================================================================
# Approximate contours to polygons + get bounding rects
contour_polygon_list = [None]*len(contours)
boundingRect_list = [None]*len(contours)
rotatedRect_list = [None]*len(contours)

#====================
for i, contour in enumerate(contours):
    # ref : https://docs.opencv.org/4.0.0/dd/d49/tutorial_py_contour_features.html

    # bounding rectangle
    epsilon = 0.1*cv2.arcLength(curve=contour, closed=True)
    contour_polygon_list[i] = cv2.approxPolyDP(curve=contour, epsilon=epsilon, closed=True)
    boundingRect_list[i] = cv2.boundingRect(array=contour_polygon_list[i])  # return (x,y, w, h)
    
    # rotated rectangle
    rect = cv2.minAreaRect(points=contour)
    box = cv2.boxPoints(box=rect)
    rotatedRect_list[i] = np.int0(box)

idx = 0
print("len(contour_polygon_list) : {}".format(len(contour_polygon_list)))
print("contour_polygon_list[{}] :\n".format(idx) +
      "{}".format(contour_polygon_list[idx]))
print("len(boundingRect_list) : {}".format(len(boundingRect_list)))
print("boundingRect_list[{}] : {}".format(idx, boundingRect_list[idx]))
print("len(rotatedRect_list) : {}".format(len(rotatedRect_list)))
print("rotatedRect_list[idx] :\n".format(idx) +
      "{}".format(rotatedRect_list[idx]))
```

### Plot

```python
rectangle_thickness = 5  # pixel
#===============================================================================
_h, _w = im.shape
_im = np.zeros(shape=(_h, _w, 3), dtype=np.uint8)
_im[:, :, :] = im[:, :, np.newaxis]

_h, _w = im.shape + 2 * np.array([rectangle_thickness, rectangle_thickness])
_im2 = np.full(shape=(_h, _w, 3), fill_value=177, dtype=np.uint8)
rectangle_thickness
_im2[rectangle_thickness:-rectangle_thickness, rectangle_thickness:-rectangle_thickness, :]= copy.deepcopy(_im)

_im3 = copy.deepcopy(_im)

print("im.shape   : {}".format(im.shape))
print("_im.shape  : {}".format(_im.shape))
print("_im2.shape : {}".format(_im2.shape))
print("_im3.shape : {}".format(_im3.shape))

#===============================================================================
def draw_boundingRect_(canvas, boundingRect, thickness, rectangle_color):
    """
    長方形を描く
    cv2.rectangle で指定するthicknessは不安定な（単純にpixel数ではない）ので以下のように自分で描画した
    bound
    """
    leftup_x   , leftup_y    = np.array(boundingRect[0:2]) + np.array([rectangle_thickness, rectangle_thickness])
    rightdown_x, rightdown_y = np.array([leftup_x-1, leftup_y-1]) + np.array(boundingRect[2:4])  # (x,y) + (w,h)
    color                    = np.array(rectangle_color)
    canvas[leftup_y-rectangle_thickness : leftup_y, leftup_x-rectangle_thickness : rightdown_x+1+rectangle_thickness, :]             = color  # 上辺
    canvas[rightdown_y+1 : rightdown_y+1+rectangle_thickness, leftup_x - rectangle_thickness : rightdown_x+1+rectangle_thickness, :] = color  # 底辺
    canvas[leftup_y-rectangle_thickness : rightdown_y+1+rectangle_thickness, leftup_x - rectangle_thickness : leftup_x, :]           = color  # 左辺
    canvas[leftup_y-rectangle_thickness : rightdown_y+1+rectangle_thickness, rightdown_x+1 : rightdown_x+1+rectangle_thickness, :]   = color  # 右辺
    return canvas

# _im  : contour_polygon
# _im2 : boundingRect
for i in range(len(contours)):
    low, high = 100, 256  # なるべく明るい色
    color = tuple([random.randint(low, high) for i in range(3)])  # scalar type should be `int()`
    cv2.drawContours(image=_im, contours=contour_polygon_list, contourIdx=i, color=color, thickness=rectangle_thickness)
    _im2 = draw_boundingRect_(canvas=_im2, boundingRect=boundingRect_list[i], thickness=rectangle_thickness, rectangle_color=color)
    cv2.drawContours(image=_im3, contours=rotatedRect_list, contourIdx=i, color=color, thickness=rectangle_thickness)

#===============================================================================
# plot
plot_array_tuple = (
    # (name, im, cmap),
    ("original", im, "gray"),
    ("drawContours", _im, None),
    ("boundingRect", _im2, None),
    ("rotatedRect", _im3, None),
)
nrows, ncols, idx  = 2, 2, 0
fig = plt.figure(figsize=(ncols*8, nrows*8))
#=======================================
plot_template(plot_array_tuple, nrows, ncols, idx, fig)
```

#### rectが傾いているのかの判定

```python
# このセルはサンプル
rect = rotatedRect_list[0]
x_min, y_min = rect.min(axis=0)
print("x_min      , y_min        : {}, {}".format(x_min, y_min))
print("x_min.dtype, y_min.dtype  : {}, {}".format(x_min.dtype, y_min.dtype))
print("rect[0]       : {}".format(rect[0]))
print("rect[0].dtype : {}".format(rect[0].dtype))
print(np.array([x_min, y_min]) in rect)
```

##### is_rotated_rect(rect)

```python
# 最後のやつは傾いているのは明らかなのでその傾きを求めてみる
rect = rotatedRect_list[-1]
print("rect :\n{}".format(rect))

def is_rotated_rect(rect):
    """
    短形が傾いているかどうか

    |-Parameters
    | rect       | numpy.array   | rectangle, 4 vertices
    |
    |-Returns
    | is_rotated | True or False |
    """
    # x座標とy座標がどちらも最小の頂点が存在すればそれは傾いていない
    x_min, y_min = rect.min(axis=0)
    if np.array([x_min, y_min]) in rect:
        return True
    return False

assert is_rotated_rect(rect=rect)
```

#### rectの傾きを求める

```python
# calculate angle
x_min_idx, y_min_idx = np.argmin(a=rect, axis=0)
# radian angle
angle_rad = np.math.atan2(rect[x_min_idx][1] - rect[y_min_idx][1],  # y
                          rect[y_min_idx][0] - rect[x_min_idx][0])  # x
# degree
angle_deg = angle_rad * 180 / np.math.pi
print("angle_rad       : {} radian".format(angle_rad))
print("type(angle_rad) : {}".format(type(angle_rad)))
print("angle_deg       : {} degree".format(angle_deg))
print("type(angle_deg) : {}".format(type(angle_deg)))
```

#### 短形の中心座標を求める

```python
# 短形の中心座標を求める
v_min, v_max = rect.min(axis=0), rect.max(axis=0)
center = v_min + (v_max - v_min) / 2
print("center          : {}".format(center))
print("center.dtype    : {}".format(center.dtype))
print("type(center[0]) : {}".format(type(center[0])))
center = tuple(map(float, center))  # numpy.float は opencv が受け付けない
print("type(center[0]) : {}".format(type(center[0])))
```

#### getRotationMatrix2D()

- [getRotationMatrix2D()](https://docs.opencv.org/4.0.0/da/d54/group__imgproc__transform.html#gafbbc470ce83812914a70abfb604f4326)
- Parameters
    - center	Center of the rotation in the source image.
    - angle	Rotation angle in degrees. Positive values mean counter-clockwise rotation (the coordinate origin is assumed to be the top-left corner).
    - scale	Isotropic scale factor.

```python
# 回転行列の取得 (同時座標)
rotate_arr = cv2.getRotationMatrix2D(center=center, angle=-angle_deg, scale=1.)
print("rotate_arr :\n{}".format(rotate_arr))
```

同時座標の概念

\begin{align*}
  \begin{bmatrix} x' \\ y' \end{bmatrix} = 
  \begin{bmatrix}
    a & b & c \\ d & e & f
  \end{bmatrix} \cdot
  \begin{bmatrix} x \\ y \\ 1 \end{bmatrix}
\end{align*}

- Refer below, if you want more details.
    - [OpenCV: Geometric Transformations of Images](https://docs.opencv.org/4.0.0/da/d6e/tutorial_py_geometric_transformations.html)
    - [画像の幾何学変換 — opencv 2.2 (r4295) documentation](http://opencv.jp/opencv-2.2/py/imgproc_geometric_image_transformations.html)
    - コンピュータグラフィックス 改訂新版 (2-1 2次元座標変換)

#### 回転の実行 cv2.warpAffine

```python
_h, _w = im.shape
im_rotated = cv2.warpAffine(src=im, M=rotate_arr, dsize=( _w, _h), flags=cv2.INTER_CUBIC)
```

#### Plot

```python
_h, _w = im.shape
_im = np.zeros(shape=(_h, _w, 3), dtype=np.uint8)
_im[:, :, :] = im[:, :, np.newaxis]
color = (0, 255, 0)  # bgr
cv2.drawContours(image=_im, contours=[rect], contourIdx=-1, color=color, thickness=5)
#===============================================================================
plot_array_tuple = (
    # (name, im, cmap),
    ("original", _im, None),
    ("RotatedIm", im_rotated, "gray"),
)
nrows, ncols, idx  = 1, 2, 0
fig = plt.figure(figsize=(ncols*8, nrows*8))
#=======================================
for (_name, _im, _cmap) in plot_array_tuple:
    idx += 1
    ax = fig.add_subplot(nrows, ncols, idx)
    ax.imshow(_im, cmap=_cmap)  # > It is ignored for RGB(A) data.
    ax.set_title(label="{}'s shape:{}".format(_name, _im.shape))
    ax.set_xlabel(xlabel="im[1]")
    ax.set_ylabel(ylabel="im[0]")
#=======================================
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```

#### 短形の幅と高さ計算

```python
# calculate edge length
x_min_idx, y_min_idx = np.argmin(a=rect, axis=0)
_, y_max_idx = np.argmax(a=rect, axis=0)

a = rect[x_min_idx] - rect[y_min_idx]
rect_width = np.sqrt(np.dot(a=a, b=a.T))
a = rect[x_min_idx] - rect[y_max_idx]
rect_height = np.sqrt(np.dot(a=a, b=a.T))

print("rect_width        : {}".format(rect_width))
print("type(rect_width)  : {}".format(type(rect_width)))
print("rect_height       : {}".format(rect_height))
print("type(rect_height) : {}".format(type(rect_height)))
```

#### 短形の切り取り

### def trim_rect_contour(image, rect)

```python
def trim_bounding_rect(image, rect):
    """
    4頂点が与えられた短形(rectangle)部分を切り抜く.
    4頂点は回転していてもよい.

    |-Parameters
    | image | numpy.array | A Input Image
    |       |             | shape(N, row, column, channel)
    |       |             | =shape(N, height, width, channel)
    | rect  | numpy.array | 4 vertices of a bounding rectangle
    |
    |-Returns
    | numpy.array | A Trimmed Image

    @pollenjp - github
    """

    #====================================
    # 短形の中心座標を求める
    # rect を囲み, かつ水平垂直な短形の upper_left, bottom_right 座標を求める
    upper_left, bottom_right = rect.min(axis=0), rect.max(axis=0)
    rect_center = upper_left + (bottom_right - upper_left) / 2

    #====================================
    # 傾いているかどうか
    if not rect.min(axis=0) in rect:  # x座標とy座標がどちらも最小の点がrectの頂点ならばそれは傾いていない
        #=================================
        # 辺の長さを求める
        rect_height, rect_width = bottom_right - upper_left  # OpenCVは左上原点であることに注意
        #=================================
        canvas = image
        canvas_center = rect_center
    else:
        #=================================
        # 辺の長さを求める
        x_min_idx, y_min_idx = np.argmin(a=rect, axis=0)
        _, y_max_idx = np.argmax(a=rect, axis=0)

        a = rect[x_min_idx] - rect[y_min_idx]
        rect_width = np.sqrt(np.dot(a=a, b=a.T))
        a = rect[x_min_idx] - rect[y_max_idx]
        rect_height = np.sqrt(np.dot(a=a, b=a.T))

        #=================================
        # 画像を回転して水平にする

        #=============
        # max(width, height)の正方形をcanvasとする　
        _h, _w, *_c = image.shape
        edge_length = max(_h, _w)
        if not _c:  # _cが空(grayscale)なら
            canvas = np.zeros(shape=(edge_length, edge_length), dtype=np.uint8)
        else:  # (color)
            canvas = np.zeros(shape=(edge_length, edge_length, _c), dtype=np.uint8)
        canvas[0:_h, 0:_w] = image  # give copied values, not reference

        #=============
        # 短形の中心がcanvasの中心になるように画像を水平移動する
        canvas_center = np.array([edge_length//2, edge_length//2])
        t_x, t_y = canvas_center - rect_center
        # 平行移動 : https://docs.opencv.org/4.0.0/da/d6e/tutorial_py_geometric_transformations.html
        M = np.array([
            [1, 0, t_x],
            [0, 1, t_y],
        ])  # 変換行列
        _h, _w, *_ = canvas.shape
        canvas = cv2.warpAffine(src=canvas, M=M, dsize=(_w, _h))

        #=============
        # 傾きの角度を計算 (calculate angle)
        x_min_idx, y_min_idx = np.argmin(a=rect, axis=0)
        angle_rad = np.math.atan2(  # radian angle (arctan(y/x))
            rect[x_min_idx][1] - rect[y_min_idx][1],  # y
            rect[y_min_idx][0] - rect[x_min_idx][0])  # x
        angle_deg = angle_rad * 180 / np.math.pi  # degree

        #=============
        # 回転させる (rotate)
        rotate_arr = cv2.getRotationMatrix2D(center=tuple(map(float, canvas_center)),  # numpy の dtype は OpenCV が受け付けない
                                             angle=-angle_deg, scale=1.)  # 回転行列
        _h, _w, *_ = canvas.shape
        canvas = cv2.warpAffine(src=canvas, M=rotate_arr, dsize=( _w, _h), flags=cv2.INTER_CUBIC)

    #================
    # trimming
    _w, _h = int(rect_width), int(rect_height)
    trimmed_rect = cv2.getRectSubPix(image=canvas,
                                     patchSize=(_w, _h),
                                     center=tuple(map(int, canvas_center))) # numpy の dtype は OpenCV が受け付けない

    return trimmed_rect
```

### Extract Bounding Box

```python
#===============================================================================
# Approximate contours to polygons + get bounding rects
contour_polygon_list = [None]*len(contours)
boundingRect_list = [None]*len(contours)
rotatedRect_list = [None]*len(contours)

#====================
for i, contour in enumerate(contours):
    # ref : https://docs.opencv.org/4.0.0/dd/d49/tutorial_py_contour_features.html

    # bounding rectangle
    epsilon = 0.1*cv2.arcLength(curve=contour, closed=True)
    contour_polygon_list[i] = cv2.approxPolyDP(curve=contour, epsilon=epsilon, closed=True)
    boundingRect_list[i] = cv2.boundingRect(array=contour_polygon_list[i])  # return (x,y, w, h)
    
    # rotated rectangle
    rect = cv2.minAreaRect(points=contour)
    box = cv2.boxPoints(box=rect)
    rotatedRect_list[i] = np.int0(box)

idx = 0
print("len(contour_polygon_list) : {}".format(len(contour_polygon_list)))
print("contour_polygon_list[{}] :\n".format(idx) +
      "{}".format(contour_polygon_list[idx]))
print("len(boundingRect_list) : {}".format(len(boundingRect_list)))
print("boundingRect_list[{}] : {}".format(idx, boundingRect_list[idx]))
print("len(rotatedRect_list) : {}".format(len(rotatedRect_list)))
print("rotatedRect_list[idx] :\n".format(idx) +
      "{}".format(rotatedRect_list[idx]))
```
